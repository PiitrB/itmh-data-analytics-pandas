## Apply() funkcia

```python
df = pd.read_csv("dataset/train.csv")
```

```python
df.head()
```

```python
df.loc[:, ["YrSold", "SalePrice"]]
```

```python
df.loc[:, ["YrSold", "SalePrice"]].apply(np.sqrt) #funkcia v parametri apply sa vola bez zatvoriek, pretoze v tomto pripade je to parameter, nie samostatna funkcia
```

```python
df.loc[:, ["YrSold", "SalePrice"]].apply(np.sqrt, axis = 1) #default apply je po stlpcoch, ak chcem riadky, tak axis = 1
```

```python
df
```

```python
def print_row(row):
    row.Id = 0
    return row
```

```python
df
```

```python
df.loc[:, ["Id", "SalePrice"]].apply(print_row, axis = 1)
```

```python
df.loc[:, ["Id", "SalePrice"]].apply(lambda row: row, axis = 1) #len ho vratim, row zostane nezmeney
```

```python
df.loc[:, ["Id", "SalePrice"]].apply(lambda row: print(f"Index is {row.name}, value is {row}"), axis = 1) 
```

## ak chcem volat viac parametrov

```python
def process_prices(row, substractor):
    row.SalePrice -= substractor
    return row
```

```python
df.loc[:, ["Id", "SalePrice"]].apply(process_prices, args=(20, ), axis = 1)
```

```python
def process_prices_1(row, sub1, sub2, sub3):
    row.SalePrice -= sub1+sub2+sub3
    return row
```

```python
df.loc[:, ["Id", "SalePrice"]].apply(process_prices_1, args=(1, 2, 3), axis = 1)
```
