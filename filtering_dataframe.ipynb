# podmienene selektovanie

```python
df.SalePrice
```

napriklad si vyfiltrujem len domy drahsie ako 175tis. Nemusim to ulozit do premennej, moze to byt aj takto: df[df.SalePrice > 175000]

```python
filter = df.SalePrice > 175000
```

```python
df[filter]
```

Mozem si vypisat kolko riadkov ma vyfiltrovany dataset

```python
sum(filter)
```

Filtrovanie podla viacerych stlpcov:
Pandas nepozna slova AND a OR ako python, ale: 
- & je AND
- | je OR
- ~ je NOT

Jednotlive podmienky __MUSIA__ byt v zatvorkach

```python
filter = (df.SalePrice > 175000) & (df.SalePrice < 223500)
```

```python
df[filter]
```

Ak to dam do NOT, tak my vylistuje vsetko, co nesplna filter.

```python
df[~filter]
```

Funkcia __IS IN__

```python
df.SalePrice.isin([175000, 250000])
```

```python
df[df.SalePrice.isin([175000, 250000])]
```
