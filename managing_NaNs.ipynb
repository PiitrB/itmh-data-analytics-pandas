```python
df.dropna(how="all") #vymaze vsetky riadky, kde kazdy stlpec ma NaN
```

```python
df.dropna(how="all", axis = 1) # to iste, len bude operovat na stlpcoch
```

```python
df.loc[:, ["Alley", "Fence"]]
```

ak by sme chceli vymazat riadky, ktore maju NaN na oboch z tych vybranych stlpcoch:

```python
df.loc[:, ["Alley", "Fence"]].isna()
```

```python
df.loc[:, ["Alley", "Fence"]].notna().apply(any, axis = 1) # funkcia apply() je funkcia vyssieho radu a aplikuje paramterom udanu funkciu na DF na ktorom bola zavolana.
# pozor, tu je param axis opacne. teda axis = 1 znamena riadky
```

```python
df.loc[df.loc[:, ["Alley", "Fence"]].notna().apply(any, axis = 1), :]
```

```python
df.loc[:, (df.dtypes == np.int64) | (df.dtypes == np.float64) ].isna().apply(any).index.values #vrati nazvy numerickych stlpcov kde sa nachadza aspon jedna NaN hodnota
```

## nahradenie NaN hodnot

```python
any(df.GarageYrBlt.isna()) #su tu NaN hodnoty
```

```python
df.loc[df.GarageYrBlt.isna(), :]
```

```python
df.loc[df.GarageYrBlt.isna(), ["GarageYrBlt"]] #vyfiltrujem si lne stlpec kde chcem nahradit nan hodnoty
```

```python
df.loc[df.GarageYrBlt.isna(), ["GarageYrBlt"]] = df.GarageYrBlt.mean()
```

```python
any(df.GarageYrBlt.isna())
```

## string hodnoty. nahradenie najpocetnejsou hodnotou
