## Ine sposoby ako pristupovat k datam a editovat ich

```python
df = pd.DataFrame(data = {"a": np.arange(1, 10), "b": np.arange(10, 19)})
df
```

```python
df["c"] = np.arange(20, 29) #pridanie stlpca. Tymto sposobom musim davat pozor na pocet prvkov. Pri Series nemusim, doplni NaN
df
```

```python
df["d"] = (df["a"] + df["b"])**2 #kombinovanie stlpcov (napr d = (a+b)**2) . pandas automaticky vie, ze vzorec ma broadcastovat na vsetky riadky, 
#pretoze poliam(vektorom) priradzujem skalarnu hodnotu **2
df
```

## Pridanie riadku:
jednou z moznosti je funckia loc

```python
df.loc[9] = [0, 1, 2, 3] # loc pozera len na hodnoty, nie na indexy. iloc nevie pridat riadok, pozera sa vyhradne na indexy a ka by som zadal cislo vacsie ako 
# najvyssi index, vrati chybu, index out of range. 
df
```

## "Mazanie" stlpcov
jeden zo sposobo je pomocou identifikatoru

```python
df[["a", "b", "d"]] #tento sposob vsak nieje vhodny ked je vela stlpcov. 
```

```python
df # povodny DF
```

Lepsi sposob

```python
rem_cols = df.columns.difference(["c"]) # funkcia columns.difference vrati iba stlpce, ktore niesu definovane v parametri...teda odcita stlpce co definujem v difference parametri
```

```python
df.loc[:, rem_cols] #loc vracia view, upravy sa premietaju do povodneho DF
```

```python
df # avsak stlpec som nevymazal naozaj, iba nezobrazil
```

```python
df.drop("c", axis=1, inplace=True) #drop je funkcia na mazanie. Axis je povinny parameter. 0 je pre riadok (default), 1 pre stlpec
# pramater inplace= je by default false. Ak dam true, tak sa data zmazu na povodnom DF. Inak opat len na kopii.
```

```python
df
```

```python
filter = (df.a >= 4) & (df.a <=6) # filtrujem hodnoty stlpca "a", nie hodnoty indexu
```

```python
df[(df.a >= 4) & (df.a <=6)] # to iste ako df[filter]
```
