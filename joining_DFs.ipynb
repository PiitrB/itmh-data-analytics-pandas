# JOINING
- inner join: vrati len stlpce/riadky (podla axis = ), ktore sa v obcho DF zhoduju, teda su v oboch DF
- outer: vrati to, co sa zhoduje a zaroven doplni co chyba

```python
df1 = pd.DataFrame({
    "first": [1,1,1,1],
    "second": [1,1,1,1],
    "third": [1,2,1,2],
}, index=[1,2,3,4])

df2 = pd.DataFrame({
    "first": [2,2,2,2],
    "second": [2,2,2,2],
    "fourth": [1,3,1,3],
}, index=[3,4,5,6])
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1, df2], join="outer", axis = 1)
```

```python
pd.concat([df1, df2], join="inner", axis = 1)
```

```python
pd.concat([df1, df2], join="inner", axis = 0)
```

```python
pd.concat([df1, df2], join="outer", axis = 0)
```
