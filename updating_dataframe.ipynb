```python
import pandas as pd
```

funkcie loc/iloc vracaju view na povodny dataset, tozn. ze ked spravim upravy na vyselektovanom "view", zmeny sa __premietnu__ aj do povodneho dataframe-u na ktorom boli funkcie loc/iloc zavolane

Ak nechcem zmenit povodny dataset a volam loc funkcie, treba pouzit funkciu copy()

Klasicke vytvaranie slice-ov pomocou hranatych zatvoriek pracuje na rozdiel od iloc/loc na kopii. Takze retazenim tychto dvoch principov vznikaju chyby, pretoze jeden princip pouziva view(upravuje base subor) a druhy vytvori copy a pracuje s nou. 

```python
df = pd.read_csv("dataset/train.csv")
```

```python
df
```

```python
df.MSZoning = 1 # nastavi hodnotu pre kazdy riadok
```

```python
df.iloc[:5, 0:4]
```

```python
hodnoty = list(range(0,1460)) #vytvori pole hodnot od 0 do 1460
```

```python
df.MSZoning = hodnoty #priradi hodnoty do stlpca MSZoning. Pripadne mozem z pola vytvorit pandas series a to potom priradit:hodnoty = pd.Series(list(range(0,1460)))
```

```python
df.loc[:4,"MSZoning"]
```

```python
df
```

```python
df[df.SalePrice > 181500] #pristupujem k domom z hodnotou vacsou ako 181500. Pozor, python vsak vytvori slice na COPY povodneho datasetu. Ak by som tento sice chcel upravit, tak povodny dataframe mi to nezmeni. Vid dalsi command
```

```python
filter = df.SalePrice > 181500
```

```python
df[filter].SalePrice = 0
```

Upravy zbehli, ale nie na povodnom dataframe, ale na jeho kopii, pretoze som pristupoval cez hranate zatvorky. 

```python
temp = df[filter]
```

```python
temp.SalePrice = 0
```

```python
temp
```

```python
df # na povodnom DF za nic nezmenilo
```

```python
df.SalePrice[filter] = 0 # na rozdiel od df[filter].SalePrice = 0 toto vsak uz povodny DF zmeni. Ale rovnako je to zly pristup, pretoze sa retazia pristupy ku DF
```

```python
df
```

## Cize dva rozdielne pristupy - oba nespravne:
- df[filter].SalePrice = 0 : zbehne s warningom, ale rovnako aj zmeni hodnoty povodneho DF
- df.SalePrice[filter] = 0 : zbehne s warningom, ale nezmeni hodnoty povodnho DF. Zmeni vsak hodnoty na kopii, na slice povodneho DF

## Spravne riesenie je pouzivat pristup loc/iloc. 
Treba sa vzdy snazit neretazit viacero podmienok/pristupov

```python
df.loc[filter, "SalePrice"] = 0
```

```python
df
```
