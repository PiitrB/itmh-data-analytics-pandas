priklad: chcem spocitat plochu ktoru zabera dom so zahradkou

```python
def total_area(group_df): #v tomto pripade posiela groupby funkcia celu grupu, preto ako pram mojej funcie bude grupa (meno je na mne) 
    return (group_df.LotFrontage + group_df.LotArea).sum() #vraciam to ako hodnotu funkcie sum(), pretoze bez toho by som dostal ako vysledo pandas series, nie celkovu sumu
```

```python
grouped.apply(total_area) #vrati sa to ako pandas series
```

```python
grouped.apply(total_area).to_frame() #vysledok premenim na DF
```

ak chceme a aby funkcia groupby vratila viacero stlpcov, je mozne dat do returnu pd.Series a nasledne pole hodnot, ktore chceme vratit

```python
def total_area(group_df): 
    return pd.Series([(group_df.LotFrontage + group_df.LotArea).sum() , group_df.LotArea.mean()], index=["Sum", "LotArea Mean"]) #napriklad chcem vratit total area ale aj priemer lotArea.
```

```python
grouped.apply(total_area)
```

da sa ten zapis vyjadrit aj pomocov LAMBDA:

```python
grouped.apply(lambda group_df: pd.Series([(group_df.LotFrontage + group_df.LotArea).sum() , group_df.LotArea.mean()], index=["Sum", "LotArea Mean"]))
```
