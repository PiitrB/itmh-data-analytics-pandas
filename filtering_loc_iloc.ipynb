## ILOC

pomocou iloc vieme pristupovat ku stlpcom a ku indexu ako ku cislam

```python
df1.iloc[[0, 1]] #iba prve dva riadky
```

```python
df1.iloc[:, [0, 1]] #iba prve dva stlpce
```

```python
df1.iloc[[0, 1], [0, 1]] #prve 2 stlpce a riadky
```

pripadne mozem pouzit syntax podobnu pythonu pri range (teda start:end). __POZOR start je INCLUDED a end je EXCLUDED__:

```python
df1.iloc[:2, :2] #vypise prve dva riadky a prve dva stlpce
```

```python
df1
```

```python
df1.iloc[-1:, :2] #viem pozuivat aj zaporne indexu, teda indexovat odkonca (posledny riadok,, prve dva stlpce)
```

```python
df1.iloc[-2:,-2:] #posledne dva riadky a posledne dva stlpce
```

## LOC

Pomocou LOC pristupujeme ku riadkom a stlpcom pomocou hodnot, nie indexov

```python
df1.loc[:2] #toto vypise vsetko, pokym nenajde riadok, kde index ma hodnotu 2...teda to nieje ako (excluded)END v iloc pripade
```

```python
df1.loc[:2, ["a", "b", 1, 2]] #v pripade stlpcov to vsak musim specifikovat explicitne. Nemozem pouzit napr :2 - teda vsetky stlpce po hodnotu 2
```

```python
df1.loc[:2, ["a", "b", 2]] #odstrani stlpec 1
```

# Prakticke priklady so suborom train.csv

```python
df = pd.read_csv("dataset/train.csv")
```

```python
df
```

```python
filter = ~((df.SalePrice > 175000) & (df.SalePrice < 223500))
```

```python
df[filter]
```

Mame zadefinovany filter na riadky, chcem vsak iba vybrany/e stlpce 

```python
df.loc[filter, "SalePrice"] # vystupom bude iba jeden stlpec, teda pandas series
```

```python
df.loc[filter, ["SalePrice", "YrSold"]] #mozem si v lubovonom poradi zadefinovat aj viacero stlpcov. Vystupom uz bude dataframe, nie seria
```

Iloc by nefungoval s tymto filtrom, pretoze zaprve nezoberie hodnoty stlpcov, a tiez neakceptuje BOOL type - co je nas filter

Alternativne si viem zobrazit filtrovane riadky a vybrane stlpce takto:

```python
df[filter][["SalePrice", "YrSold"]] #ale odporuca sa radsej pouzit loc
```
